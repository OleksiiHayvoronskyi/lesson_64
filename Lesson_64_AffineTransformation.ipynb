## За допомогою графічного редактора визначте точки відповідності (кути дошки) на зображеннях **chess_1** та **chess_2**, що додані до репозиторію лекції.
## Проведіть гомографічне перетворення для приведення зображення **chess_2** до вигляду базового зображення **chess_1**.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Reading and displaying first image.
img_chess_1 = cv2.imread('chess_1.png')
img_chess_2 = cv2.imread('chess_2.png')

# Creating a figure for a subplot.
fig = plt.figure(figsize=(20, 16))

# Subplot and title of the first image.
ax1 = fig.add_subplot(221)  # top left side.
ax1.title.set_text('Chess 1')

# Subplot and title of the second image.
ax2 = fig.add_subplot(222)  # top right side.
ax2.title.set_text('Chess 2')

# Display the subplot.
ax1.imshow(img_chess_1)
ax2.imshow(img_chess_2)
ax1.axis('off')
ax2.axis('off')
plt.show()
```

```python
# Source coordinates.
output_img = np.array([[211, 81], [449, 73], [213, 219], [443, 231]])

# Destination coordinates.
final_img = np.array([[221, 313], [217, 68], [380, 347], [470, 117]])

# output_img = np.array([[211, 81], [449, 73], [213, 219], [443, 231]])
#
# # Destination coordinates.
# final_img = np.array([[221, 313], [217, 68], [380, 347], [470, 117]])
```

```python
# Homography implementation.
h, status = cv2.findHomography(output_img, final_img)

im_out = cv2.warpPerspective(img_chess_1, h,
                             (img_chess_2.shape[1],
                             img_chess_2.shape[0]))

plt.figure(figsize=(10, 8))
plt.title('Homography image', fontsize=14, color='blue')
plt.imshow(im_out)
plt.show()
```
